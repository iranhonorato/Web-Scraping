```python
import pandas as pd
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service 
from selenium.webdriver.common.by import By 
import time
```

```python
def inicializa_driver():
    # A classe Service é usada para iniciar uma instância do Chrome Webdriver
    service = Service()

    # webdriver.ChromeOptions é usada para definir a preferência para o browser do chrome
    options = webdriver.ChromeOptions()

    # Inicia-se a instância do Chrome Webdriver com as definidas "options" e "service"
    return webdriver.Chrome(service=service, options=options)
```

```python
driver = inicializa_driver()

url = "https://repositorio.insper.edu.br/home"
driver.get(url)
time.sleep(0.75)



def fazer_busca(driver, nome):
    links = []

    # Preenche o campo de busca
    busca_repositorio = driver.find_element(By.CSS_SELECTOR, 'input[data-test="search-box"]')
    busca_repositorio.click()
    busca_repositorio.clear()
    busca_repositorio.send_keys(nome)
    time.sleep(1)

    # Clica no botão de busca
    botao_pesquisar = driver.find_element(By.CSS_SELECTOR, 'button[data-test="search-button"]')
    botao_pesquisar.click()
    time.sleep(2)

    while True:
        try:
            pesquisas = driver.find_elements(By.CSS_SELECTOR, "a.item-list-title")
            lista = [link.get_attribute("href") for link in pesquisas]
            links.extend(lista)

            proxima_pagina = driver.find_element(By.CSS_SELECTOR, 'a.page-link[aria-label="Next"]')

            # Verifica se está desabilitado
            if "disabled" in proxima_pagina.get_attribute("class"):
                break

            proxima_pagina.click()
            time.sleep(2)

        except:
            break
    return links
```

```python
nome = "Sérgio Lazzarini"
pesquisador = {} 
pesquisador[nome] = fazer_busca(driver, nome)

print(pesquisador)
```

```python
linhas = []

for trabalhos in pesquisador.values():
    for trabalho in trabalhos:
        linhas.append({"Nome": "Sérgio Lazzarini", "Trabalho": trabalho})


# metricis.to_excel("Metricis.xlsx", index=False)
faltantes = pd.DataFrame(linhas)
faltantes
```

```python
classificacao = []
data = [] 

driver = inicializa_driver()

for url in faltantes["Trabalho"]:
    driver.get(url)
    time.sleep(0.75)


    try:
        tipo_doc = driver.find_element(
        "xpath",
        "//div[h2[text()='Tipo de documento']]//span").text

        classificacao.append(tipo_doc)
    except:
        classificacao.append("Não foi possível identificar o tipo do trabalho")



    try:    
        data_elem = driver.find_element(
        "xpath",
        "//div[h2[text()='Data']]//span").text
        
        data.append(data_elem)
    except:
        data.append("Não foi possível identificar a data do trabalho")
```

```python
faltantes.insert(loc=2, column="Data", value=data)
faltantes.insert(loc=3, column="Categoria", value=classificacao)
faltantes
```

```python
autores_lista = []

for url in faltantes["Trabalho"]:

    driver.get(url)
    time.sleep(0.75)

    try:
        autores_elements = driver.find_elements(
            "xpath",
            "//div[h2[text()='Autores']]//a | //div[h2[text()='Autores']]//span"
        )
        
        autores_texto = [autor.text.lower().strip() for autor in autores_elements if autor.text.lower().strip()]
        autores_lista.append(", ".join(autores_texto))

    except:
        autores_lista.append("Não foi possível identificar os autores")
```

```python
faltantes.insert(loc=4, column="Autores", value=autores_lista)
faltantes
```

```python
faltantes.to_excel("Sergio_Lazzarini.xlsx", index=False)
```
